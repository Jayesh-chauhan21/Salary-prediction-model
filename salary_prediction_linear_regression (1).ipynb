**Salary Prediction** <br>
--> In this tutorial we will use simple linear regression to predict the salary of a person based on their work of experience. <br>
--> <b>Linear Regession</b> is very basic model of supervised Machine Learning <br>
It uses Y = m*x + c formulae <br> where Y is prediction variablr, m is slope X is constant on which model Y is to be predicted"X depends on Y", and C is constant


```python
#importing reqired dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
#loading data to our notebook
data = pd.read_csv('Salary Data.csv')
data.head()
```

```python
#getting statistical info about our data
data.describe()
```

```python
#hetting information about data
data.info()
```

```python
#visualizing data in graphical mode to get better understanding
#here we used scatter plot
plt.scatter(x=data['YearsExperience'], y=data['Salary'] , color='red', marker='D')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
#defined LinearRegression from scikit learn
#then we fitted to linear regression model
reg = linear_model.LinearRegression()
reg.fit(data[['YearsExperience']], data['Salary'])
```

```python
#here is the output of our data i.e, predicted salary(Y) aacording to experience(X)
reg.predict([[12]])
```
